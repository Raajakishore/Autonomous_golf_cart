```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('stop_sign.jpg')

plt.imshow(img)
```

```python
plt.imshow(img)
print(img.shape[0])
plt.plot(445,165,'.') #tr 440, 110
plt.plot(455,235,'.') #br 440, 180
plt.plot(285,185,'.') #bl 280, 180
plt.plot(285,115,'.') #tl 280, 110
```

```python
def warp(img):
    
    img_size = (491, 541)
    
    src = np.float32([
        [445, 165],
        [455, 235],
        [285, 185],
        [285, 115]
    ])
    dst = np.float32([
        [440, 110],
        [440, 180],
        [280, 180],
        [280, 110]
    ])
    
    M = cv2.getPerspectiveTransform(src, dst)
    
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    return warped
    
```

```python
wraped_img = warp(img)

f, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))

ax1.set_title('source image')
ax1.imshow(img)
ax2.set_title('warped image')
ax2.imshow(wraped_img)
```
